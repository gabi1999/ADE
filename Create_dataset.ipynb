```python
import pandas as pd
import numpy as np
import os
from sqlalchemy import create_engine
```

```python
def read_file(filename):
    data = pd.read_csv(filename, delimiter = ",", skiprows = 7, header = None)
    return data
```

```python
df = pd.DataFrame(columns = ["User", "Day", "Latitude", "Longitude", "Hora"])
df
```

```python
read_file("./Dataset/Data/000/Trajectory/20081023025304.plt")
```

```python
users = os.listdir("./Dataset/Data")
for user in users:
    if user != ".DS_Store":
        filepath = "./Dataset/Data/" + user + "/Trajectory/"
        files = os.listdir(filepath)
        #print(files)
        for file in files:
            df_data = read_file(filepath + "/" + file)  
            
            df.loc[len(df)] = [user, df_data.at[0, 5], df_data[0].values, df_data[1].values, df_data[6].values]
  
df
```

```python
df = df.sort_values(by = ["User", "Day"])
df = df.reset_index()
df
```

```python
itineration_1 = pd.DataFrame(columns = ["Latitude", "Longitude", "Hora"])

itineration_1["Latitude"] = df.at[0, "Latitude"]
itineration_1["Longitude"] = df.at[0, "Longitude"]
itineration_1["Hora"] = df.at[0, "Hora"]
itineration_1
```
